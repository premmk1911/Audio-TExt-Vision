```python
import pyttsx3
import speech_recognition as sr
from gtts import gTTS
import os
import socket

def speak(text):
    engine = pyttsx3.init()
    engine.say(text)
    engine.runAndWait()
    
    
def display_and_speak_instructions(instructions):
    print(instructions)
    speak(instructions)
    
    

def text_to_speech():
    display_and_speak_instructions("Please enter the text you want to convert to speech.")
    text_to_speak = input("Your text: ")
    speak(text_to_speak)
    tts = gTTS(text=text_to_speak, lang='en', slow=False)
    s="Do you want to save the audio? "
    display_and_speak_instructions(s)
    sch=input("Y/N : ")
    if (sch=='Y' or sch=='y'):
        sav=("Enter a name for the audio file without extension :")
        display_and_speak_instructions(sav)
        file_name = input("Enter the file name : ")
        # Check if the file already exists
        while os.path.exists(f"{file_name}.mp3"):
            display_and_speak_instructions("File already exists. Please choose a different name.")
            file_name = input("Enter a new name: ")
            # Convert text to speech and save the audio file
        tts.save(f"{file_name}.mp3")
        # Play the saved audio file
        #os.system(f"start {file_name}.mp3")
        display_and_speak_instructions(f"Audio file saved as {file_name}.mp3 in Documents.")
        
        

def speech_to_text():
    # Check for internet connectivity
    display_and_speak_instructions("\nMake sure to use working microphone and internet connection\n\t\t")
    try:
        socket.create_connection(("www.google.com", 80))
    except OSError:
        display_and_speak_instructions("No Internet connection,\n\nPlease connect to the internet and run the program")
        exit()
    display_and_speak_instructions("Please speak the text you want to convert to written form.")
    
    recognizer = sr.Recognizer()
    with sr.Microphone() as source:
        print("Say something...\n\n\n\t\tüó£Ô∏è))üëÇüèº  \n\n\nPress 'ctrl+c' to end this process")
        try:
            audio = recognizer.listen(source, timeout=10)
        except sr.WaitTimeoutError:
            print("No audio detected for 10 seconds. Ending program.")
            return
    try:
        text = recognizer.recognize_google(audio)
        print("You said:", text)
    except sr.UnknownValueError:
        print("Could not understand audio")
    except sr.RequestError as e:
        print(f"Could not request results; {e}")

    
    
# Main menu
while True:
    
    menu_instructions = "MENU:\nPress 1 for Text to Speech\nPress 2 for Speech to Text\nPress 3 for Exit"
    display_and_speak_instructions(menu_instructions)

    choice = input("Enter your choice (1, 2, or 3): ")

    if choice == '1':
        
        text_to_speech()
        
    elif choice == '2':
        speech_to_text()
        
    elif choice == '3':
        display_and_speak_instructions("Exiting program. Goodbye!")
        break
    else:
        display_and_speak_instructions("Invalid choice. Please enter 1, 2, or 3.")
```

```python

#importing libraries
import pyttsx3
import speech_recognition as sr
from gtts import gTTS
import os
import socket

#to speak the text
def speak(text,speed,lan):
    engine = pyttsx3.init()
    engine.setProperty('rate',speed)
    engine.setProperty('voice',lan)
    engine.say(text)
    engine.runAndWait()

#to display and speak the instructions
def display_and_speak_instructions(instructions):
    print(instructions)
    speak(instructions,210,'en')
    
    
#to convert text to speech in 2 languages
def text_to_speech():
    display_and_speak_instructions("Enter the language \npress 1 for english\npress 2 for hindi\nBy default the text is converted in english language. ")
    ch_lan=int(input("Enter 1or2 : "))
    display_and_speak_instructions("Please enter the text you want to convert to speech.")
    #takeing user input for the language
    text_to_speak = input("Your text: ")
    
    #checking the user choice of language
    if (ch_lan==2):    # for hindi language
        engine = pyttsx3.init()
        engine.setProperty('rate',180)  #setting the speed of speech in words per min.
        engine.setProperty('voice','hi') #setting the language
        engine.say(text_to_speak)    #works in system which has hindi speech recognitition
        engine.runAndWait()
        save_audio(text_to_speak,'hi')   #calling the function to save the audio
        
    #By default the text is converted in english language.
    else:
        speak(text_to_speak,180,'en')   #calling speak function to read the text by setting setting the speech in words per min.
        save_audio(text_to_speak,'en')  #calling the function to save the audio
        
        
 #to save the text to speech audio
def save_audio(text_to_speak,lan):
    s="Do you want to save the audio? "
    display_and_speak_instructions(s) #asking to save the file or not
    sch=input("Y/N : ")
    
    #checking user input
    if (sch=='Y' or sch=='y'):
        tts = gTTS(text=text_to_speak, lang=lan, slow=True) #setting the speed and language
        sav=("Enter a name for the audio file without extension :") 
        display_and_speak_instructions(sav)  #asking name of file to save from user
        file_name = input("Enter the file name : ")
        # Check if the file already exists
        while os.path.exists(f"{file_name}.mp3"):
            display_and_speak_instructions("File already exists. Please choose a different name.")
            file_name = input("Enter a new name: ")
            # Convert text to speech and save the audio file
        tts.save(f"{file_name}.mp3")
        # Play the saved audio file
        #os.system(f"start {file_name}.mp3")
        display_and_speak_instructions(f"Audio file saved as {file_name}.mp3 in Documents.")
        
        

#to converting the speech to text
def speech_to_text():
    # Check for internet connectivity
    display_and_speak_instructions("\nMake sure to use working microphone and internet connection\n\t\t")
    try:
        socket.create_connection(("www.google.com", 80))
    except OSError:  
        #if no internet connected then it displays the message and exit 
        display_and_speak_instructions("No Internet connection,\n\nPlease connect to the internet and run the program")
        exit()
        
    display_and_speak_instructions("Please speak the text you want to convert to written form.")
    recognizer = sr.Recognizer()
    with sr.Microphone() as source:
        print("Say something...\n\n\n\t\tüó£Ô∏è))üëÇüèº  \n\n\nPress 'ctrl+c' to end this process")
        try:
            audio = recognizer.listen(source, timeout=10)
        #if no audio recognised for 10 sec then displays WaitTimeoutError
        except sr.WaitTimeoutError:
            print("No audio detected for 10 seconds. Ending program.")
            return
    try:
        text = recognizer.recognize_google(audio) #converting the speech to text
        print("You said:\n'", text.upper(),"'")  #text is displayed in upper
        
    #when audio is not clear
    except sr.UnknownValueError: 
        print("Could not understand audio")
    
    except sr.RequestError as e:
        print(f"Could not request results; {e}")

    
    
# Main menu
while True:
    
    #displaying the instructions
    menu_instructions = "MENU:\nPress 1 for Text to Speech\nPress 2 for Speech to Text\nPress 3 for Exit"
    display_and_speak_instructions(menu_instructions)

    choice = input("Enter your choice (1, 2, or 3): ")

    if choice == '1':
        text_to_speech()
        
    elif choice == '2':
        speech_to_text()
        
    elif choice == '3':
        display_and_speak_instructions("Exiting program. Goodbye!")
        break
    else:
        display_and_speak_instructions("Invalid choice. Please enter 1, 2, or 3.")
        
        
        
```
